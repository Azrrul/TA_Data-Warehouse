```python
data = pd.read_excel("Dataset_TA.xlsx")
data
```

```python
checkup = []
tahun = []
gender = []
kota = []
for i,j in data.iterrows():
    if j[1] not in checkup:
        checkup.append(j[1])
        
    k = -1
    for index, value in enumerate(checkup):
        if value == j[1]:
            k = index+1
            break
        
    if[j[0], j[2], j[3], k] not in tahun:
        tahun.append([j[0], j[2], j[3], k])
```
